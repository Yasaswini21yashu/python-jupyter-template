### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to generate x and y values with error handling
def generate_data():
    try:
        # Generate x values from 0 to 10
        x = np.linspace(0, 10, 100)
        
        # Compute y values as the square of x
        y = x ** 2
        
        # Check if the data is valid
        if len(x) == 0 or len(y) == 0:
            raise ValueError("Generated data is empty.")
        
        return x, y
    
    except Exception as e:
        print(f"Error during data generation: {e}")
        return None, None

# Function to create the plot with error handling
def create_plot(x, y):
    try:
        if x is None or y is None:
            raise ValueError("Invalid data: Cannot plot empty or None values.")
        
        # Create the plot
        plt.plot(x, y, label='y = x²')

        # Customize the plot
        plt.title('Line Plot of y = x²')
        plt.xlabel('x values')
        plt.ylabel('y values')
        plt.grid(True)
        plt.legend()

        # Display the plot
        plt.show()
    
    except Exception as e:
        print(f"Error during plotting: {e}")

# Generate data and create the plot
x, y = generate_data()
create_plot(x, y)
```
